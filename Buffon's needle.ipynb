```python
import random
import math as m
import numpy as np
import matplotlib.pyplot as plt

def solve(n):
    c = 0
    for i in range(n):
        theta = random.random()*180
        x = random.random()*6
        y = random.random()*6
        y_cordinates = list(map(lambda x: round(x, 3), np.arange(y-0.5*m.sin(theta*m.pi/180),y+0.5*m.sin(theta*m.pi/180),0.001)))
        if any(x in y_cordinates for x in [0,4,2,6]):
            c+=1
    return c/n
l = []
p = 100
for i in range(1,1000,p):
    l.append(abs(1/m.pi-solve(i)))
    # print(solve(i),'difference with 1/pi :',abs(1/m.pi-solve(i)))
plt.plot(range(1,1000,p),l)
plt.xticks(range(0,1000,p))
# plt.xscale(np.log())
plt.show()
```
